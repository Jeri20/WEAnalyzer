```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.linkextractors import LinkExtractor
from scrapy.spiders import CrawlSpider, Rule

class WebsiteSpider(CrawlSpider):
    name = "website_spider"
    allowed_domains = []  # Will be dynamically set
    start_urls = []
    rules = [Rule(LinkExtractor(), callback='parse_page', follow=True)]

    custom_settings = {
        'CONCURRENT_REQUESTS': 32,
        'CONCURRENT_REQUESTS_PER_DOMAIN': 16,
        'CONCURRENT_REQUESTS_PER_IP': 16,
        'DOWNLOAD_DELAY': 0.1,
        'RANDOMIZE_DOWNLOAD_DELAY': False,
        'COOKIES_ENABLED': False,
        'RETRY_ENABLED': False,
        'AUTOTHROTTLE_ENABLED': False,
        'DNS_RESOLVER': 'scrapy.resolver.CachingAsyncResolver',
        'DOWNLOAD_HANDLERS': {
            'http': 'scrapy.core.downloader.handlers.http2.H2DownloadHandler',
            'https': 'scrapy.core.downloader.handlers.http2.H2DownloadHandler',
        },
        'USER_AGENT': 'scrapy-user-agents',
        'LOG_LEVEL': 'ERROR',
    }

    def parse_page(self, response):
        yield {
            'url': response.url,
            'title': response.xpath('//title/text()').get(),
            'headings': response.xpath('//h1//text() | //h2//text() | //h3//text()').getall(),
            'meta_description': response.xpath('//meta[@name="description"]/@content').get(),
            'content': ' '.join(response.xpath('//p//text()').getall())
        }

def crawl_website(url):
    domain = url.split("//")[-1].split("/")[0]
    process = CrawlerProcess(settings=WebsiteSpider.custom_settings)
    WebsiteSpider.allowed_domains = [domain]
    WebsiteSpider.start_urls = [url]
    process.crawl(WebsiteSpider)
    process.start()
```

```python
from playwright.sync_api import sync_playwright
import json
import asyncio

async def crawl_website(url, max_pages=50):
    crawled_data = []
    visited_urls = set()
    to_visit = [url]

    async with sync_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context()

        while to_visit and len(crawled_data) < max_pages:
            page_url = to_visit.pop(0)

            if page_url in visited_urls:
                continue

            visited_urls.add(page_url)
            page = await context.new_page()

            try:
                await page.goto(page_url, timeout=60000)

                title = await page.title()
                content = await page.content()

                headings = await page.locator('h1, h2, h3').all_inner_texts()
                meta_description = await page.locator('meta[name="description"]').get_attribute('content')

                crawled_data.append({
                    'url': page_url,
                    'title': title,
                    'headings': headings,
                    'meta_description': meta_description,
                    'content': content
                })

                # Extract links for further crawling
                links = await page.locator('a').evaluate_all('elements => elements.map(el => el.href)')
                to_visit.extend([link for link in links if link.startswith(url) and link not in visited_urls])

            except Exception as e:
                print(f"Failed to crawl {page_url}: {e}")
            finally:
                await page.close()

        await browser.close()

    with open('output.json', 'w') as f:
        json.dump(crawled_data, f, indent=4)

    return crawled_data
```

```python
from transformers import pipeline
import json

ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
intent_pipeline = pipeline("text-classification", model="bhadresh-savani/bert-base-uncased-emotion")

def analyze_content(file_path='output.json'):
    with open(file_path, 'r') as f:
        data = json.load(f)

    for page in data:
        content = page['content']
        entities = ner_pipeline(content)
        intent = intent_pipeline(content[:512])
        page['entities'] = entities
        page['intent'] = intent[0]['label']

    with open('analysis.json', 'w') as f:
        json.dump(data, f, indent=4)
```

```python
import gradio as gr

def run_tool(url):
    crawl_website(url)
    analyze_content()
    with open('analysis.json', 'r') as f:
        analysis_data = json.load(f)
    return analysis_data

gr.Interface(
    fn=run_tool,
    inputs=gr.Textbox(label="Website URL"),
    outputs=gr.JSON(label="Analysis Output"),
    title="Fast Website Entity & Intent Analyzer",
    description="Enter a website URL to crawl, extract content, and analyze entities and search intent using Hugging Face models, with speed-optimized crawling."
).launch()
```
